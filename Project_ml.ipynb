````python
import ast
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# change the above as list
#  want the o/p as ['Action','Adventure','Fantasy','Sci-fi']
def convert(text):
  l = []
  for i in ast.literal_eval(text):
    l.append(i['name'])
  return l
````

````python
movies['genres'] = movies['genres'].apply(convert)
````

````python
movies.head()
````

````python
movies['keywords'] = movies['keywords'].apply(convert)
````

````python
def convert3(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      l.append(i['name'])
      counter += 1
    else:
      break
  return l
````

````python
movies['cast'] = movies['cast'].apply(convert)
````

````python
def fetch_director(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      l.append(i['name'])
      break
  return l
````

````python
movies['crew'] = movies['crew'].apply(fetch_director)
````

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset (replace 'movie_data.csv' with your actual dataset filename)
# movie_data = pd.read_csv('movie_data.csv')

# Extract movie tags from the DataFrame
corpus = new_df['tags']

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Convert movie tags into TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

# Define the number of folds for cross-validation
k = 5  # Number of folds

# Initialize KFold cross-validator
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Initialize lists to store evaluation scores for each fold
evaluation_scores = []

# Iterate over each fold
for fold, (train_index, test_index) in enumerate(kf.split(tfidf_matrix), 1):
    print(f"Fold {fold}:")
    # Split the TF-IDF matrix into training and testing sets for the current fold
    X_train, X_test = tfidf_matrix[train_index], tfidf_matrix[test_index]

    # Compute cosine similarity between movies in the training set
    similarity_matrix_train = cosine_similarity(X_train, X_train)

    # Compute cosine similarity between movies in the testing set and movies in the training set
    similarity_matrix_test = cosine_similarity(X_test, X_train)

    # Calculate the mean similarity score for the testing set
    mean_similarity = similarity_matrix_test.mean()

    # Store the evaluation score for the current fold
    evaluation_scores.append(mean_similarity)

    print(f"Mean Similarity Score for Fold {fold}: {mean_similarity:.4f}\n")

# Plot the evaluation scores
plt.figure(figsize=(8, 6))
plt.plot(range(1, k+1), evaluation_scores, marker='o', linestyle='-')
plt.title('Performance Across Folds')
plt.xlabel('Fold')
plt.ylabel('Mean Similarity Score')
plt.xticks(range(1, k+1))
plt.grid(True)
plt.show()
````
